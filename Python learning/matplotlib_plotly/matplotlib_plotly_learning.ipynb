```python
import pandas as pd

# Load the dataset
df = pd.read_csv('laliga_24_25_games.csv')
```

```python
pd.set_option('display.max_columns', 500)  
pd.set_option('display.max_rows', 500)  
```

```python
# Display the first few rows
print(df.shape)
df.head()
```

### Histogram

```python
# Drop missing values in FTHG (if any)
home_goals = df['FTHG'].dropna()
```

```python
# Plot the distribution of full-time home goals
plt.figure(figsize=(8, 5))
plt.hist(home_goals)

# Add labels and title
plt.title('La Liga 2024/25: Distribution of Home Goals')
plt.xlabel('Goals Scored by Home Team')
plt.ylabel('Number of Matches')

# Show plot
plt.show()
```

```python
# Plot the distribution of full-time home goals
plt.figure(figsize=(8, 5))
plt.hist(home_goals, bins=range(0, int(home_goals.max()) + 2), color='skyblue', edgecolor='black', align='left')

# Add labels and title
plt.title('La Liga 2024/25: Distribution of Home Goals')
plt.xlabel('Goals Scored by Home Team')
plt.ylabel('Number of Matches')
plt.xticks(range(0, int(home_goals.max()) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Plot histogram
plt.hist(home_goals,
         bins=range(0, int(home_goals.max()) + 2),
         color='skyblue',
         edgecolor='black',
         align='left')

# Calculate and plot the mean line
mean_goals = home_goals.mean()
plt.axvline(mean_goals, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_goals:.2f}')

# Labels and grid
plt.title('La Liga 2024/25: Distribution of Home Goals')
plt.xlabel('Goals Scored by Home Team')
plt.ylabel('Number of Matches')
plt.xticks(range(0, int(home_goals.max()) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show legend and plot
plt.legend()
plt.show()
```

### Bar Chart

```python
# Group by home team and sum their home goals
top_home_scorers = df.groupby('HomeTeam')['FTHG'].sum().sort_values(ascending=False).head(10)
```

```python
# Plot
plt.figure(figsize=(10, 5))
top_home_scorers.plot(kind='bar', color='seagreen', edgecolor='black')

# Labels and title
plt.title('Top 10 Home Scoring Teams - La Liga 2024/25')
plt.xlabel('Team')
plt.ylabel('Total Home Goals')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### Line chart

```python
# Make sure date is in datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Drop rows where date is missing
df = df.dropna(subset=['Date'])

# Group by date and sum home goals
goals_by_date = df.groupby('Date')['FTHG'].sum()
```

```python
# Plot
plt.figure(figsize=(10, 5))
goals_by_date.plot(kind='line', marker='o', color='steelblue')

# Labels and styling
plt.title('La Liga 2024/25: Home Goals Over Time')
plt.xlabel('Match Date')
plt.ylabel('Total Home Goals')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Group by date
goals_by_date = df.groupby('Date')['FTHG'].sum()

# Rolling average (5 days)
rolling_avg = goals_by_date.rolling(window=5).mean()

# Plot both
plt.figure(figsize=(10, 5))
plt.plot(goals_by_date, label='Daily Goals', color='steelblue', marker='o', linestyle='-')
plt.plot(rolling_avg, label='5-Day Rolling Avg', color='red', linewidth=2)

# Labels and legend
plt.title('La Liga 2024/25: Home Goals Over Time (Smoothed)')
plt.xlabel('Match Date')
plt.ylabel('Total Home Goals')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Show only 1 tick every 7 days
# plt.xticks(ticks=goals_by_date.index[::7], rotation=45)

plt.tight_layout()
plt.show()
```

### Scatter plot

```python
# Create X and Y
x = df['HS']     # Home Shots
y = df['FTHG']   # Full Time Home Goals
```

```python
# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.6, color='teal', edgecolor='black')

# Labels and title
plt.title('La Liga 2024/25: Home Shots vs Home Goals')
plt.xlabel('Shots by Home Team')
plt.ylabel('Goals Scored by Home Team')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#### Showcasing outliers
import numpy as np

# Plot the main scatter
plt.figure(figsize=(9, 6))
plt.scatter(x, y, alpha=0.6, color='teal', edgecolor='black', label='Match')

# Outliers: many shots but 0 goals
outliers = df[
    ((df['HS'] >= 10) & (df['FTHG'] == 0)) |
    ((df['HS'] > 20) & (df['FTHG'] < 2))
]

plt.scatter(outliers['HS'], outliers['FTHG'], color='orange', edgecolor='black', label='Outliers (10+ shots, 0 goals)', zorder=3)

# Labels, grid, legend
plt.title('La Liga 2024/25: Home Shots vs Goals (with Trend + Outliers)')
plt.xlabel('Home Shots')
plt.ylabel('Home Goals')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

### Pie chart

```python
# Count each match outcome
result_counts = df['FTR'].value_counts()

# Map result codes to labels
labels = {'H': 'Home Win', 'D': 'Draw', 'A': 'Away Win'}
results = [labels[r] for r in result_counts.index] #Were converting short result codes ('H', 'D', 'A') into full text for the pie labels
```

```python
# Plot
plt.figure(figsize=(6, 6))

plt.pie(result_counts, labels=results, autopct='%1.1f%%', colors=['lightgreen', 'gold', 'salmon'], startangle=90) # show % inside each slice (1 decimal)
plt.title('La Liga 2024/25: Match Outcomes')
plt.axis('equal')  

plt.tight_layout()
plt.show()
```

# Plotly

```python
# Plotly Express is the high-level, quick-and-easy interface
# Use this for fast, simple charts with just 1 line of code (like px.bar, px.histogram, etc.)
import plotly.express as px

# Plotly Graph Objects is the low-level, fully customizable interface
# Use this when you need full control (layout, annotations, multi-trace charts, etc.)
import plotly.graph_objects as go
```

### Histogram

```python
import plotly.express as px

fig = px.histogram(df, x='FTHG', nbins=int(df['FTHG'].max() + 1),
                   title='Distribution of Home Goals (La Liga 2024/25)',
                   labels={'FTHG': 'Goals Scored by Home Team'},
                   color_discrete_sequence=['skyblue'])

fig.update_layout(bargap=0.1)
fig.show()
```

### Bar chart

```python
top_teams = df.groupby('HomeTeam')['FTHG'].sum().sort_values(ascending=False).head(10).reset_index()

fig = px.bar(top_teams, x='HomeTeam', y='FTHG',
             title='Top 10 Home Scoring Teams',
             labels={'FTHG': 'Total Home Goals', 'HomeTeam': 'Team'},
             color_discrete_sequence=['seagreen'])

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
import plotly.express as px

# Step 1: Group and sort the data
top_teams = df.groupby('HomeTeam')['FTHG'].sum().sort_values(ascending=False).head(10).reset_index()

# Step 2: Create the bar chart
fig = px.bar(
    top_teams,
    x='HomeTeam',
    y='FTHG',
    title='Top 10 Home Scoring Teams - La Liga 2024/25',
    labels={'HomeTeam': 'Team', 'FTHG': 'Total Home Goals'},
    color_discrete_sequence=['seagreen']
)

# Step 3: Make it clean
fig.update_layout(
    xaxis_tickangle=-45,
    template='simple_white'
)

# Step 4: Show the interactive chart
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Prepare data
goals_by_date = df.groupby('Date')['FTHG'].sum().reset_index() #Full Time Home Goals (FTHG) 
goals_by_date['RollingAvg'] = goals_by_date['FTHG'].rolling(window=5).mean() #it calculates the rolling averageof the FTHG column over the last 5 rows.



# Build the figure
fig = go.Figure()

# Line 1: Actual daily goals
fig.add_trace(go.Scatter(
    x=goals_by_date['Date'],
    y=goals_by_date['FTHG'],
    mode='lines+markers',
    name='Daily Goals',
    line=dict(color='steelblue')
))

# Line 2: Rolling average
fig.add_trace(go.Scatter(
    x=goals_by_date['Date'],
    y=goals_by_date['RollingAvg'],
    mode='lines',
    name='5-Day Rolling Avg',
    line=dict(color='red', width=2, dash='solid')
))

# Layout styling
fig.update_layout(
    title='La Liga 2024/25: Home Goals Over Time (Smoothed)',
    xaxis_title='Match Date',
    yaxis_title='Total Home Goals',
    hovermode='x unified',
    template='simple_white'
)

fig.show()
```

### Scatter Plot

```python
# Define main x and y
x = df['HS']     # Home Shots
y = df['FTHG']   # Full-Time Home Goals

# Define outliers
outliers = df[
    ((df['HS'] >= 10) & (df['FTHG'] == 0)) |
    ((df['HS'] > 20) & (df['FTHG'] < 2))
]

# Create figure
fig = go.Figure()

# 1 Main scatter plot (normal matches)
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='markers',
    name='Match',
    marker=dict(color='teal', size=8, opacity=0.6, line=dict(color='black', width=1))
))

# 2 Outliers (highlighted in orange)
fig.add_trace(go.Scatter(
    x=outliers['HS'],
    y=outliers['FTHG'],
    mode='markers',
    name='Outliers (10+ shots, <2 goals)',
    marker=dict(color='orange', size=10, line=dict(color='black', width=1))
))

# Layout & styling
fig.update_layout(
    title='La Liga 2024/25: Home Shots vs Goals (with Outliers)',
    xaxis_title='Home Shots',
    yaxis_title='Home Goals',
    template='simple_white',
    legend=dict(x=0.7, y=1),
    hovermode='closest'
)

fig.show()
```

### Pie Chart

```python
# Step 1: Count outcomes
result_counts = df['FTR'].value_counts().reset_index()
result_counts.columns = ['ResultCode', 'Count']

# Step 2: Map short codes to full names
label_map = {'H': 'Home Win', 'D': 'Draw', 'A': 'Away Win'}
result_counts['ResultLabel'] = result_counts['ResultCode'].map(label_map)
```

```python
# Step 3: Create the pie chart
fig = px.pie(
    result_counts,
    names='ResultLabel',
    values='Count',
    title='La Liga 2024/25: Match Outcomes',
    color='ResultLabel',
    color_discrete_map={
        'Home Win': 'lightgreen',
        'Draw': 'gold',
        'Away Win': 'salmon'
    },
    hole=0  # Set to 0.3 or 0.5 for a donut chart
)

# Optional: Make it round and clean
fig.update_traces(textinfo='percent+label')
fig.update_layout(showlegend=True, template='simple_white')

fig.show()
```
